## Concatenating the all extracted csv file from the following the regions

- Afghanistan
- Bangladesh
- Indonesia
- Veitnam

### Import necessary python packages

```python
import pandas as pd
import glob
```

### Locating the extracted csv file names into a list

```python
extension = 'csv'
all_filenames = [i for i in glob.glob('C:\\Users\\GM\\Desktop\\!Code!\\CDRI\\Alcoholics Anonymous\\Data Extraction\\#Dataset\\*.{}'.format(extension))]
```

### combine all files in the list and export to a new csv file

```python
combined_csv = pd.concat([pd.read_csv(f, engine='python') for f in all_filenames ])

combined_csv.to_csv("Final.csv", index=False, encoding='utf-8')
```
